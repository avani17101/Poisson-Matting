```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
# import plotly.express as px
import os
IMAGES_FOLDER_PATH = "../img"
```

```python
img1 = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'img1.jpg'))
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1_gray = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'img1.jpg'),0)

tri1 = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'tri1.jpg'),0)

print(img1.shape)
print(tri1.shape)
```

```python
plt.figure(figsize = (20,15))

plt.subplot(131)
plt.imshow(img1)
plt.title('Original Image')
plt.subplot(132)
plt.imshow(tri1,cmap='gray')
plt.title('Trimap')
plt.subplot(133)
plt.imshow(img1_gray,cmap='gray')
plt.title('Grayscale')


plt.show()
```

```python
img2 = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'img2.jpg'))
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2_gray = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'img2.jpg'),0)

tri2 = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'tri2.jpg'),0)

print(img2.shape)
print(tri2.shape)
```
